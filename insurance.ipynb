```python
# Now we will find the numeric columns to make plot
numeric_columns = ['age', 'bmi', 'children','charges']
numeric_columns
for col in numeric_columns:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col],kde=True,bins=20)
    #kde=True  adds a smooth curve showing the overall pattern
    #bins=20 divides your data into 20 groups → 20 bars
```

```python
# Doing the same plots but in subplots
fig,axes = plt.subplots(1, 4, figsize =(20,6))
for i,col in enumerate(numeric_columns):
    i
    col
    sns.histplot(df[col], ax=axes[i],kde=True, bins=20)
    axes[i].set_title(col)
plt.tight_layout()
plt.show()
```
